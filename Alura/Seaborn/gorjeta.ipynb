# Importando base de dados

```python
import pandas as pd
```

```python
dados = pd.read_csv('tips.csv')
dados
```

# Tradução

```python
dados.columns
```

```python
renomear = {'total_bill': 'valor_da_conta', 'tip': 'gorjeta', 'dessert': 'sobremesa',
            'day': 'dia_da_semana', 'time': 'hora_do_dia', 'size': 'total_de_pessoas'}
renomear
```

```python
gorjetas.sobremesa.unique()
```

```python
gorjetas.sobremesa.map(sim_nao)
```

```python
gorjetas.dia_da_semana.unique()
```

```python
gorjetas.dia_da_semana = gorjetas.dia_da_semana.map(dias)
gorjetas
```

```python
gorjetas.hora_do_dia.unique()
```

```python
hora = {'Dinner': 'Jantar', 'Lunch': 'Almoço'}
```

```python
gorjetas.hora_do_dia = gorjetas.hora_do_dia.map(hora)
gorjetas
```

## Teste de hipótese

**H<sup>null</sup>**

> **A distribuição da taxa da gorjeta é a mesma nos dois grupos**

**H<sup>null</sup>**

> **A distribuição da taxa da gorjeta NÃO é a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
sobremesa=gorjetas.query('sobremesa == "Sim"').porcentagem
```

```python
sem_sobremesa=gorjetas.query('sobremesa == "Não"').porcentagem
```

```python
r=ranksums(sobremesa,sem_sobremesa)
```

```python
print('O valor do p-value é {}'.format(r.pvalue))
```

> **A distribuição da taxa da gorjeta é a mesma nos dois grupos**

# Teste de hipótese

**H<sup>null**
> **A distribuição do valor da conta é igual no sábado eno domingo**

**H<sup>alt**
> **A distribuição do valor da conta NÃO é igual no sábado eno domingo**

```python
valor_conta_domingo=gorjetas.query('dia_da_semana == "Domingo"').valor_da_conta
```

```python
valor_conta_sabado=gorjetas.query('dia_da_semana == "Sábado"').valor_da_conta
```

```python
r2 = ranksums(valor_conta_domingo,valor_conta_sabado)
print(f'O valor do p-value é {round(r2.pvalue,2)}')
```

**H<sup>null**
> **A distribuição do valor da conta é igual no sábado eno domingo**

## Teste de Hipótese

```python
gorjetas.groupby(by='hora_do_dia').mean().drop('total_de_pessoas',axis=1)
```

**H<sup>null**
> **A distribuição do valor da conta é igual no jantar eno almoço**

> **A distribuição do valor da conta NÃO é igual no jantar eno almoço**

```python
r3 = ranksums(gorjetas.query('hora_do_dia == "Jantar"').valor_da_conta,gorjetas.query('hora_do_dia == "Almoço"').valor_da_conta)
print(f'O valor do p-value é {round(r3.pvalue,5)}')
```

**H<sup>alt**
> **A distribuição do valor da conta NÃO é igual no jantar eno almoço**

```python
r3_2 = ranksums(gorjetas.query('hora_do_dia == "Jantar"').porcentagem,gorjetas.query('hora_do_dia == "Almoço"').porcentagem)
print(f'O valor do p-value é {round(r3_2.pvalue,5)}')
```

**H<sup>null**
> **A distribuição do percentual de gorjeta é igual no jantar eno almoço**
